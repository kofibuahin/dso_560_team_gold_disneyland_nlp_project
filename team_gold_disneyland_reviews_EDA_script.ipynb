```python
import pandas as pd

data = [['HongKong', 4.204158, 9620], ['California',4.405339, 19406], ['Paris', 3.960088, 13630]]
meanrating = pd.DataFrame(data, columns=['Location', 'Mean Rating', 'Reviews'])

meanrating
```

```python
import plotly.express as px
fig = px.bar(meanrating, x='Location', y='Mean Rating', color = meanrating.Location)
fig.show()
```

```python
fig = px.pie(meanrating, values='Reviews', names='Location', title = 'Pie Chart of Review Count by Location')
fig.show()
```

```python
import plotly.express as px
fig = px.histogram(disney, x='Rating', y='Rating', color = 'Branch',barmode='group', histfunc='count')
fig1 = fig.update_layout(yaxis_title = "Count of Reviews")
fig1.show()
```

```python
#Used this as a reference: https://www.geeksforgeeks.org/python-pandas-split-strings-into-two-list-columns-using-str-split/
newdata = disney.Year_Month.str.split('-', n=1, expand=True)
disney['Year'] = newdata[0]
disney['Month'] = newdata[1]
disney.head()
```

```python
edadf = disney[disney.Year!='missing']
edadf
```

```python
import plotly.express as px
fig = px.histogram(edadf, x='Year', y='Rating', color = 'Branch',barmode='group', histfunc='avg')
fig1 = fig.update_layout(yaxis_title = "Average of Rating")
fig1.show()
```
